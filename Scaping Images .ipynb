```python
import pandas as pd
import numpy as np

import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://en.wikipedia.org/wiki/Yoga'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
images = soup.find_all('img', {'src':re.compile('.jpg')})
```

```python
images
```

```python
count = 0
for image in images: 
    print(image['src']+'\n')
    count = count + 1
    
print("There are", count, " Number of images in the given link")
```
